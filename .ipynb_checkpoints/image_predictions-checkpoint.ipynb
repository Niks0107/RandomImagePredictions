```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras.applications import vgg16
from keras.preprocessing.image import img_to_array,load_img
```

```python
currentDirectory = os.getcwd()
print(currentDirectory)
```

```python
imageDir = os.path.join(currentDirectory,"Data")
imageDir
```

```python
for img in os.listdir(imageDir):
    print(img)
```

```python
model=vgg16.VGG16()
```

```python
image = load_img(os.path.join(imageDir,"parrot.jpg"),target_size=(224,224))
imagearray = img_to_array(image)
```

```python
imagearray.shape
```

```python
imagearray = imagearray.reshape(1,imagearray.shape[0],imagearray.shape[1],imagearray.shape[2])
imagearray = vgg16.preprocess_input(imagearray)
pred = model.predict(imagearray)
```

```python
vgg16.decode_predictions(pred)
```

```python
plt.imshow(image)
plt.title(vgg16.decode_predictions(pred)[0][0][1])
```

```python
def modelPrediction(imageDir):
    for images in os.listdir(imageDir):
        image = load_img(os.path.join(imageDir,images), target_size = (224,224))
        imagearray = img_to_array(image)
        imagearray = imagearray.reshape(1,imagearray.shape[0],imagearray.shape[1],imagearray.shape[2])
        imagearray = vgg16.preprocess_input(imagearray)
        pred = model.predict(imagearray)
        name = vgg16.decode_predictions(pred)[0][0][1]
        plt.imshow(image)
        plt.title(name)
        plt.show()
```

```python
modelPrediction(imageDir)
```
